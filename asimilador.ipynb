```
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

# Lista de frases
sentences = [
    'Los bucles se utilizan para repetir acciones.',
    'Las condiciones permiten realizar acciones dependiendo de ciertas situaciones.',
    'Las funciones agrupan un conjunto de instrucciones.',
    'Las listas almacenan colecciones de elementos.',
    'Los errores pueden ocurrir durante la ejecución del programa.',
    'La programación orientada a objetos organiza el código en clases y objetos.',
    'Los comentarios son útiles para explicar el código a otras personas.',
    'Las variables almacenan información importante.'
]

# Cómputo de embeddings
embeddings = model.encode(sentences)
```

```
embeddings
```

```

cosine_scores = util.cos_sim(embeddings, embeddings)
# Encuentra las parejas con los puntajes de similitud del coseno más altos
pairs = []

# Recorre la matriz de puntajes de similitud coseno
for i in range(len(cosine_scores)-1):
    for j in range(i+1, len(cosine_scores)):

        # Agrega la pareja a la lista con su índice y puntaje
        pairs.append({'index': [i, j], 'score': cosine_scores[i][j]})

# Ordena las parejas por puntaje en orden descendente
pairs = sorted(pairs, key=lambda x: x['score'], reverse=True)

# Imprime las 5 mejores parejas
for pair in pairs[0:5]:
    i, j = pair['index']

    # Imprime las oraciones con los índices de la pareja
    print("{}-->{}".format(sentences[i], sentences[j]))
```
